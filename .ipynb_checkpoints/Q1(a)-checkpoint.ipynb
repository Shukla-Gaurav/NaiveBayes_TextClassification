```python
import utils as utility
import numpy as np
import pickle

def get_modal(docs,stemming):
    stars = [0]*5
    category_count = [0]*5

    class_frequency = {}
    count = 0
    all_words = []
    
    for doc in docs:
        count = count + 1
        if(stemming):
            words = utility.getStemmedDocuments(doc['text'])
        else:
            words = doc['text'].split()

        star = int(doc['stars'])
        stars[star-1] += 1
        category_count[star-1] += len(words)

        for word in words:
            if word not in class_frequency:
                all_words.append(word)
                class_frequency[word] = [1]*5
            class_frequency[word][star-1] += 1

    #print(class_frequency)
    #print(all_words)

    m = count
    vocab_size = len(all_words)
    for i in range(5):
        category_count[i] += vocab_size

    frequency_matrix = [class_frequency[i] for i in all_words]

    parameter_matrix = np.array(frequency_matrix)/np.array(category_count)

    phai_y = np.array(stars)/m
 
    parameters = [all_words,parameter_matrix,phai_y]
    #write the object into the file
    if(stemming):
        file_name = "probability_stemming"
    else:
        file_name = "probability_withoutstemming"
    # open the file for writing
    obj_writer = open(file_name,'wb') 
    
    pickle.dump(parameters,obj_writer)   
    obj_writer.close()
    print('done')
    return 

docs =  utility.json_reader('../train.json')
get_modal(docs,False)
get_modal(docs,True)
```

```python
import math
def get_the_object(file_name):
    # we open the file for reading
    obj_reader = open(file_name,'rb')  
    parameters = pickle.load(obj_reader)  
    obj_reader.close()
    return parameters


def get_predicted_class(review,stemming,parameters):
    all_words = parameters[0]
    prob_matrix = parameters[1]
    phai_y=parameters[2]
    
    print(prob_matrix)
    vocab_size = len(all_words)
    if(stemming):
        words = utility.getStemmedDocuments(review)
    else:
        words = review.split()
        
    sum_of_logs = [math.log(prob_y) for prob_y in phai_y]
    print(sum_of_logs)
    for word in words:
        if word not in all_words:
            for i in range(5):
                sum_of_logs[i] += math.log(1/vocab_size)
        else:
            ind = all_words.index(word)
            print(ind)
            for i in range(5):
                sum_of_logs[i] += math.log(prob_matrix[ind][i])
    
    max_value = sum_of_logs[0]
    max_ind = 0
    for i in range(1,5):
        if(sum_of_logs[i] > max_value):
            max_value = sum_of_logs[0]
            max_ind = i
    return max_ind+1

def compute_testdata_accuracy_stemming():
    docs =  utility.json_reader('../test.json')
    count = 0
    correct_prediction = 0
    parameters = get_the_object("probability_stemming")
    for doc in docs:
        count += 1
        prediction = get_predicted_class(doc['text'],True,parameters)
        star = int(doc['stars'])
        
        if(star == prediction):
            correct_prediction += 1
    return correct_prediction/count

def compute_testdata_accuracy():
    docs =  utility.json_reader('../test.json')
    count = 0
    correct_prediction = 0
    parameters = get_the_object("probability_withoutstemming")
    for doc in docs:
        count += 1
        prediction = get_predicted_class(doc['text'],False,parameters)
        star = int(doc['stars'])
        
        if(star == prediction):
            correct_prediction += 1
    return correct_prediction/count
    
print(compute_testdata_accuracy())
print(compute_testdata_accuracy_stemming())
```
